```python
# set depedencies
import pandas as pd
import numpy as np
import os
```

```python
#load assesment file
Hills_Values = os.path.join ("../Resources/HillsboroughAsmt.csv")

#create dataframe
Hills_Values_df = pd.read_csv(Hills_Values, dtype={'PHY_ZIPCD':'str'})

Hills_Values_df.head()
```

```python
#pick necessary columns
Home_Values_df = Hills_Values_df [['SEQ_NO','PARCEL_ID','JV','NBRHD_CD','CENSUS_BK','PHY_ADDR1','PHY_CITY','PHY_ZIPCD']]
Home_Values_df.head()
```

```python
#remove any null values 
home_value1_df =Home_Values_df.dropna()
home_value1_df.head()
```

```python
#remame columns
renamed_df = home_value1_df.rename(columns={"JV": "PropertyValue",
                                            "PHY_ADDR1": "PropAddress",
                                            "PHY_CITY":"City",
                                            "PHY_ZIPCD":"ZipCode"})
renamed_df.head()
```

```python
#create data frame for specific zip code


Zip_df = renamed_df.loc[(renamed_df["ZipCode"]=="33613")]
Zip_df.head()
```

```python
#export to csv 
Zip_df.to_csv("../Resources/Hill_Prop_Value.csv")
```
