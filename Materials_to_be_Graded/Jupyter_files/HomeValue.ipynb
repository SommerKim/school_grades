```python
# set depedencies
import pandas as pd
import numpy as np
import os
```

```python
#load assesment file
Hills_Values = os.path.join ("../Resources/GB/HillsboroughAsmt.csv")

#create dataframe
Hills_Values_df = pd.read_csv(Hills_Values, dtype={'PHY_ZIPCD':'str','TOT_LVG_AREA':'str','ACT_YR_BLT':'str'})

Hills_Values_df.head()
```

```python
#pick necessary columns
Home_Values_df = Hills_Values_df [['SEQ_NO','PARCEL_ID','JV','NBRHD_CD','CENSUS_BK','PHY_ADDR1','PHY_CITY','PHY_ZIPCD','TOT_LVG_AREA','ACT_YR_BLT']]
Home_Values_df.head()
```

```python
#remove any null values 
home_value1_df =Home_Values_df.dropna()
home_value1_df.head()
```

```python
#remame columns
renamed_df = home_value1_df.rename(columns={"JV": "PropertyValue",
                                            "PHY_ADDR1": "PropAddress",
                                            "PHY_CITY":"City",
                                            "PHY_ZIPCD":"ZipCode",
                                             "TOT_LVG_AREA":"BldgSqFt",
                                           "ACT_YR_BLT": "YearBuilt"})
renamed_df.head()
```

```python
renamed_df.to_csv("../Resources/GB/All_Hill_Prop_Value.csv")
```

```python
#create data frame for specific zip code


Zip_df = renamed_df.loc[(renamed_df["ZipCode"]=="33613")|(renamed_df["ZipCode"]=="33556")]

Zip_df.head()
```
